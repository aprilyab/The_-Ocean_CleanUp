make the appropraite numeric value

```python
loaded_data["MPW (metric tons year -1)"] = pd.to_numeric(
    loaded_data["MPW (metric tons year -1)"].astype(str).str.replace("'", "").str.replace(",", ""),
    errors="coerce"
)
```

creating a new column with the name 'plastic_contribution' based on the value of "MPW (metric tons year -1)"

```python
loaded_data['plastic_contribution'] = loaded_data["MPW (metric tons year -1)"].apply(
    lambda x: "0" if x > 6008 else "1"
)
```
